```python
#data cleaning steps
for x in data.columns:
    for i in range(len(data)):
        # Remove the word starting with @
        data.loc[i,x] = re.sub(r'@[A-Za-z0-9]+', ' ', str(data.loc[i,x]))
  
        #Remove URL links
        data.loc[i,x]  = re.sub('https?://[A-Za-z0-9./]+',' ',data.loc[i,x] ) 
        data.loc[i,x]  = re.sub('http?://[A-Za-z0-9./]+',' ',data.loc[i,x] )

        # Converting to Lowercase
        data.loc[i,x]  = data.loc[i,x] .lower()
        
        #Remove the new line characters
        data.loc[i,x] = re.sub(r"\t|\n|\r", " ", data.loc[i,x] , flags=re.I)
    
        #Remove punctuation
        data.loc[i,x] = re.sub(r"[,@\#:'?\.$%_!()]", " ", data.loc[i,x] , flags=re.I)

        #Remove digits
        data.loc[i,x] = re.sub(r"\d", "", data.loc[i,x] )

        # remove all single characters
        data.loc[i,x] = re.sub(r'\s+[a-zA-Z]\s+', ' ', data.loc[i,x] )

        # Substituting multiple spaces with single space
        data.loc[i,x]  = re.sub(r'\s+', ' ', data.loc[i,x] , flags=re.I)
    
```

```python
## split into words, create tokens
from nltk.tokenize import word_tokenize
for x in data.columns:
    for i in range(len(data)):
        data.loc[i,x] = word_tokenize(str(data.loc[i,x]))
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
for x in data.columns:
    for i in range(len(data)):
        data.loc[i,x] = [word for word in data.loc[i,x] if not word in stop_words]
```

```python
data.head(15)
```

```python
nltk.download('wordnet')
```

```python
# Lemmetization of words (it will change the word in the base form)
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for x in data.columns:
    for i in range(len(data)):
        data.loc[i,x] = [lemmatizer.lemmatize(word,pos='a') for word in data.loc[i,x]]
        data.loc[i,x] = [lemmatizer.lemmatize(word,pos='v') for word in data.loc[i,x]]
        data.loc[i,x] = [lemmatizer.lemmatize(word,pos='n') for word in data.loc[i,x]] 
```

```python
data.head(15)
```

```python
# remove remaining tokens that are not alphabetic
for x in data.columns:
    for i in range(len(data)):
        data.loc[i,x] = [word for word in data.loc[i,x] if word.isalpha()]
```
